**Merging the two data frames**

```python
df.dropna(inplace=True)
df.head()
```

**Recommended system with one example of "Shahr Shik BMW keychain"**

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


df = df.copy()


new_df = df[["product_title_en", "category_keywords", "brand_name_en"]].copy()
new_df["CombinedFeatures"] = (
    new_df["product_title_en"] + " " + new_df["category_keywords"] + " " + new_df["brand_name_en"]
)

#new_df.info()

new_df["CombinedFeatures"] = new_df["CombinedFeatures"].fillna("")


vectorizer = CountVectorizer(stop_words="english")
features_vectors = vectorizer.fit_transform(new_df["CombinedFeatures"])

fav = "Shahr Shik BMW keychain"


if fav in new_df["product_title_en"].values:
    idx = new_df.index.get_loc(new_df[new_df["product_title_en"] == fav].index[0])
else:
    raise ValueError(f"'{fav}' not found in product_title_en")


if idx >= features_vectors.shape[0]:
    raise IndexError(f"Index {idx} is out of range for feature vectors with shape {features_vectors.shape}")


similarities = cosine_similarity(features_vectors, features_vectors[idx])


products = np.c_[new_df["product_title_en"], similarities]


num_recommended = 3


recommended = sorted(products, key=lambda product: float(product[1]), reverse=True)


print(np.array(recommended)[1:num_recommended + 1, 0])

```
